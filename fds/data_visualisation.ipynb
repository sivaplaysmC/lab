```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'year': range(2010,2021),
    'job_postings': [150, 300, 450, 600, 800, 1200, 1600, 2100, 2700, 3400, 4200]
}

df = pd.DataFrame(data)
plt.plot(df["year"], df["job_postings"], marker='*')
plt.title("Data science job postings")
plt.xlabel("year")
plt.ylabel("job_postings")

plt.show()

print(df.head(100))
```
